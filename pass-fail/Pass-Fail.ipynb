```python
import pandas as pd
import numpy as np
from  sklearn.linear_model import LogisticRegression
url ='/content/drive/MyDrive/student-pass-fail-data.csv'

namelist = ['Self_Study_Daily','Tutorials_Monthly','Pass_Or_Fail']
iris = pd.read_csv(url , header=0,names = namelist)

iris1 = np.array(iris)
X = iris1[:,0:2]
Y = iris1[:,2]

model = LogisticRegression()
model.fit(X,Y)

prediction_input = np.array([[5, 37]])
print(model.predict(prediction_input)[0])
```
